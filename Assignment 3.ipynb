```python
driver=webdriver.Chrome(r"C:\Users\Jigar\Downloads\chromedriver_win32")
```

```python
driver.get("https://www.amazon.in/")
```

```python
tags=driver.find_element(By.XPATH,'//div[@class="nav-search-field "]/input')
tags.send_keys('guitar')
```

```python
search=driver.find_element(By.XPATH,'html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search.click()
time.sleep(3)
```

```python
guitar_url=[]
```

```python
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        guitar_url.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]') 
    
        
```

```python
brand=[]
for url in guitar_url:
    driver.get(url)
    time.sleep(2)
    try:
        BN=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[46]/div/table/tbody/tr[1]')
        brand.append(BN.text)
    except NoSuchElementException:
        brand.append('-')
```

```python
len(brand)
```

```python
brand
```

```python
product_name=[]
for url in guitar_url:
    driver.get(url)
    time.sleep(2)
    try:
        PN=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        product_name.append(PN.text)
    except NoSuchElementException:
        product_name.append('-')
```

```python
len(product_name)
```

```python
product_name
```

```python
price=[]
for url in guitar_url:
    driver.get(url)
    time.sleep(2)
    try:
        PT=driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center"]')
        price.append(PT.text)
    except NoSuchElementException:
        price.append('Not present')
```
